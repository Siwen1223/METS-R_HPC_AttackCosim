```python
# copy the saved simulation to the data folder
import shutil

# go to the output folder and fetch the latest CARLAT05 folder, then copy the test.zip to the data folder
import os
output_folder = os.path.join(os.getcwd(), "output")
latest_folder = max([os.path.join(output_folder, d) for d in os.listdir(output_folder)], key=os.path.getmtime)   
shutil.copy(os.path.join(latest_folder, "test.zip"), os.path.join(os.getcwd(), "data", "test.zip"))
```
